# Scrubbing #

1st: create a vector "variable1" that will be a list of all the changes from one timepoint to the next

For example, if the globalsignal.txt file is:   1, 5, 6, 7, 8, 9, 10, 15, 16, 17, 18,
variable1 will be: variable1 = 5-1,  6-5, 7-6, 9-7 ... 18-17 
which equals       variable1 = 4, 1, 1, 1, 1, 1, 5, 1, 1, 1

2nd: calculate the mean and std deviation of variable1. 

3rd: find the outliers. That is; calculate how many values in variable1 fall below mean-(2*std) and above mean+(2*std)

4th: specify the position of the outlier in the list. In our example globalsignal.txt, the outliers will probably be the jumps from 1 to 5 and from 10 to 15, so positions 2 and 8. Number 5 is pos 2 and number 15 is pos 8.

The final outcome should be 1 the number of outliers in every subject and 2. a list of outlier positions for each subject. 


### 1. Creating differences vectors ###

```python
import numpy as np
import pandas as pd 
import sys
import os
from glob import glob
import matplotlib.pyplot as plt
```

```python
data_path = '/Users/rotemgatenyo/Desktop/Technion/Semester 8/Final Project/Scrubbing/Control'

data_dirs = os.path.join(data_path, "*")
all_dirs = glob(data_dirs)

control = []

for a_dir in all_dirs:
    dir_path = os.path.join(a_dir, "*")
    all_file_paths = (glob(dir_path))
    for a_file_path in all_file_paths:
        temp_data_df = pd.read_csv(a_file_path)
        temp_data_df = pd.DataFrame.transpose(temp_data_df)
        splitted_file = a_file_path.split(os.sep)
        temp_data_df['filename'] = splitted_file[9]
        control.append(temp_data_df)

control = pd.concat(control)
control = control.reset_index()
control = control.set_index('filename')
control.columns = range(control.columns.size)

control
```

```python
data_path = '/Users/rotemgatenyo/Desktop/Technion/Semester 8/Final Project/Scrubbing/Dyslexia'

data_dirs = os.path.join(data_path, "*")
all_dirs = glob(data_dirs)

dyslexia = []

for a_dir in all_dirs:
    dir_path = os.path.join(a_dir, "*")
    all_file_paths = (glob(dir_path))
    for a_file_path in all_file_paths:
        temp_data_df = pd.read_csv(a_file_path)
        temp_data_df = pd.DataFrame.transpose(temp_data_df)
        splitted_file = a_file_path.split(os.sep)
        temp_data_df['filename'] = splitted_file[9]
        dyslexia.append(temp_data_df)

dyslexia = pd.concat(dyslexia)
dyslexia = dyslexia.reset_index()
dyslexia = dyslexia.set_index('filename')
dyslexia.columns = range(dyslexia.columns.size)

dyslexia
```

```python
control = control.apply(pd.to_numeric)
control_diff = control.diff(axis=1)

dyslexia = dyslexia.apply(pd.to_numeric)
dyslexia_diff = dyslexia.diff(axis=1)
```

## 2. Calculate the mean and std deviation ##

```python
control_des = control_diff.apply(pd.DataFrame.describe, axis=1)
control_des
```

```python
dyslexia_des = dyslexia_diff.apply(pd.DataFrame.describe, axis=1)
dyslexia_des
```

## 3. Finding the outliers. ##
We defined the outliers as the variables outside the range [Q1 - 1.5IQR,Q3 + 1.5IQR] where IQR = Q3-Q1

```python
# control_des['IQR'] = control_des['75%']- control_des['25%']
# control_des['low range'] = control_des['25%'] -1.5*control_des['IQR'] 
# control_des['top range'] = control_des['75%'] +1.5*control_des['IQR'] 

# dyslexia_des['IQR'] = dyslexia_des['75%']- dyslexia_des['25%']
# dyslexia_des['low range'] = dyslexia_des['25%'] -1.5*dyslexia_des['IQR'] 
# dyslexia_des['top range'] = dyslexia_des['75%'] +1.5*dyslexia_des['IQR'] 

control_des['low range'] = control_des['mean'] -2*control_des['std'] 
control_des['top range'] = control_des['mean'] +2*control_des['std'] 

dyslexia_des['low range'] = dyslexia_des['mean'] -2*dyslexia_des['std'] 
dyslexia_des['top range'] = dyslexia_des['mean'] +2*dyslexia_des['std'] 

control_des
```

```python
dyslexia_des
```

## 4. specify the position of the outlier in the list ##

```python
control_diff_t = pd.DataFrame.transpose(control_diff)
control_des_t = pd.DataFrame.transpose(control_des)

control_outlier = control_diff_t[((control_diff_t < (control_des_t.loc['low range'])) |(control_diff_t > (control_des_t.loc['top range'])))]
control_outlier = pd.DataFrame.transpose(control_outlier)
control_outlier
```

```python
num_outliers = control_outlier.apply(lambda x: x.count(), axis=1)
print(num_outliers)
num_outliers.sum()
```
